```python
import pandas as pd
import numpy as np
from datetime import datetime
```

 # CUSTOMER_DATA 

```python
file_path = "D:\GUVI\Project\Dataspark project\Dataset\Customers.csv"
df_customers = pd.read_csv(file_path,encoding='latin1')
df_customers
```

```python
# remove duplicate
df_customers = df_customers.drop_duplicates()
```

```python
# check null value in the dataset
df_customers.isnull().sum()
```

```python
# drop state code column (State code and state provide same information)
df_customers = df_customers.drop(columns = ['State Code','Zip Code'],axis = 1)
df_customers
```

```python
# check datatypes of the column 
df_customers.info()
```

```python
# change Birthday column datatype to datatime
df_customers['Birthday'] = pd.to_datetime(df_customers['Birthday'])
df_customers['Birthday']
```

```python
df_customers.info()
```

```python
# calculate age from the Brithday column

def calculate_age(birthdate):
    today = datetime.today()
    age = today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
    return age

df_customers['Age'] = df_customers['Birthday'].apply(calculate_age)

def age_segment(age):
    if age < 25:
        return 'young adult'
    elif age < 55:
        return 'middle age'
    else:
        return 'senior citizen'

# Apply the function to the 'age' column
df_customers['Age Category'] = df_customers['Age'].apply(age_segment)
    
```

```python
df_customers.columns = df_customers.columns.str.lower()
df_customers.columns = df_customers.columns.str.replace(" ","_")
```

```python
df_customers
```

```python
csv_file_name = "df_customers.csv"
df_customers.to_csv(csv_file_name, index=False)
```

# PRODUCT_DATA

```python
file_path = "D:\GUVI\Project\Dataspark project\Dataset\Products.csv"
df_products = pd.read_csv(file_path)
df_products
```

```python
# remove duplicate
df_products = df_products.drop_duplicates()
```

```python
# check null value in the dataset
df_products.isnull().sum()
```

```python
# information of the columns in the datasets 
df_products.info()
```

```python
df_products['Unit Cost USD'] = (
    df_products['Unit Cost USD']
    .str.replace('[^\d.,]', '', regex=True)  # Remove all non-numeric characters except ',' and '.'
    .str.replace(',', '', regex=True)       # Remove commas
    .str.strip()                            # Strip any extra spaces
    .replace('', '0')                       # Replace empty strings with '0' (if any remain)
    .astype(float)                          # Convert to float
)

df_products['Unit Price USD'] = (
    df_products['Unit Price USD']
    .str.replace('[^\d.,]', '', regex=True)  # Remove all non-numeric characters except ',' and '.'
    .str.replace(',', '', regex=True)       # Remove commas
    .str.strip()                            # Strip any extra spaces
    .replace('', '0')                       # Replace empty strings with '0' (if any remain)
    .astype(float)                          # Convert to float
)

```

```python
# add new column
df_products['Profit Margin'] = df_products['Unit Price USD'] - df_products['Unit Cost USD']
df_products['Profit Margin']
```

```python
df_products
```

```python
df_products.info()
```

```python
df_products.columns = df_products.columns.str.lower()
df_products.columns = df_products.columns.str.replace(" ","_")
```

```python
csv_file_name = "df_products.csv"
df_products.to_csv(csv_file_name, index=False)
```
