```python
import pandas as pd
import pymysql
```

# SALES TABLE

```python
file_path = "D:\GUVI\Project\Dataspark project\df_sales.csv"
df_sales = pd.read_csv(file_path)
df_sales
```

```python
# MySQL connection

myconnection = pymysql.connect(
    host="localhost", 
    user="root", 
    passwd="Iyyappan1*", 
    database="dataspark"
)
```

```python
df_sales.columns
```

```python
# Create sales table
myconnection.cursor().execute("""CREATE TABLE IF NOT EXISTS sales (order_number INT NOT NULL,
                                                                         line_item INT,
                                                                         order_date DATE,
                                                                         customer_key INT,
                                                                         store_key INT,
                                                                         product_key INT,
                                                                         quantity INT,
                                                                         currency_code VARCHAR(3))""")
# Insert Values in sales table
for i in range(len(df_sales)):
    myconnection.cursor().execute(f"insert into sales values {tuple(df_sales.iloc[i])}")
    myconnection.commit()

```

```python
# Create customer table

myconnection.cursor().execute("""CREATE TABLE IF NOT EXISTS customers (customerkey INT PRIMARY KEY,
                                                                      gender VARCHAR(6),
                                                                      name VARCHAR(40),
                                                                      city VARCHAR(40),
                                                                      state VARCHAR(40),
                                                                      country VARCHAR(39),
                                                                      continent VARCHAR(30),
                                                                      birthday DATE,
                                                                      age INT,
                                                                      age_category VARCHAR(20)
                                                                         )""")
# Insert Values in customer table
for i in range(len(df_customers)):
    myconnection.cursor().execute(f"insert into customers values {tuple(df_customers.iloc[i])}")
    myconnection.commit()

```
